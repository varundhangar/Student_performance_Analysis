```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.metrics import mean_absolute_error , mean_squared_error,r2_score
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
data=pd.read_csv("student_performance_1000_rows.csv")
df_label=data.copy()

le=LabelEncoder()


df_label["Gender"]=le.fit_transform(df_label['Gender'])
x=df_label[["Study Hours Per Week","Assignment Average","Quiz Average","Gender","Midterm Score","Attendance (%)","Participation Score","Project Score"]]
y=df_label["Final Score"]
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
model=LinearRegression()
model.fit(X_train ,y_train)
predicted_scores=model.predict(X_test)


mse = mean_squared_error(y_test,predicted_scores)
mae = mean_absolute_error(y_test,predicted_scores)
rmse = np.sqrt(mse)
r2=r2_score(y_test,predicted_scores)




print("mean squared value" ,round(mse,2))
print("mean absoluteeroor",round(mae,2))
print("root means value",round(rmse,2))
print("round value of this project ",round(r2,2))#closer to 1 =better

#histogram
plt.figure(figsize=(10,6))
plt.hist(df_label["Final Score"],bins=30 ,color='skyblue',edgecolor='black')
plt.grid()
plt.title("distribution of final exm")
plt.xlabel("final_exam_score")
plt.ylabel("number of student")
plt.show()

# scatter graph
plt.figure(figsize=(10,6))
plt.scatter(y_test,predicted_scores,color='blue',label='Actual Scores')

plt.grid(True)
plt.title("model prediction vs actaual scores")
plt.xlabel("study Hours per week ")
plt.ylabel("number of student")
plt.show()


predicted_new_scores=model.predict([[18,70,60,1,98,77.11,80,78]])
print(f"predicted final scores for Hours is {predicted_new_scores}")

          





```
